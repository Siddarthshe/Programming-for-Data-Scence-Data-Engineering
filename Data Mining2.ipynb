```python
import pandas as pd
import numpy as np
data = pd.read_csv('auto.csv')
data
```

```python
null_data = data[data.isnull().any(axis=1)]
```

```python
data = data.replace('?' , np.NaN)
data = data.fillna(0)
data
```

```python
data.columns[-1]
```

```python
X = data.drop('mpg',axis=1)
y = data['mpg']
```

```python
dummies = pd.get_dummies(X['car name']).rename(columns=lambda x: 'car name_x' + str(x))

X = pd.concat([X, dummies], axis=1)
X = X.drop(['car name'], axis=1)
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train_org, X_test_org, y_train, y_test = train_test_split(X, y, random_state = 0)

scaler = StandardScaler() # for regression - standard because the way it is going to scale the data bring the outliers close to the data;classification,pca - minmaxscaler
X_train = scaler.fit_transform(X_train_org)
X_test = scaler.transform(X_test_org)
```

```python
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()
lreg.fit(X_train, y_train)
print(lreg.score(X_train, y_train)) # score - r-square
print(lreg.score(X_test, y_test))
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(random_state= 0, max_iter = 100000, learning_rate = 'optimal', penalty = 'l1')
sgd_reg.fit(X_train, y_train)
print(sgd_reg.score(X_train, y_train))
print(sgd_reg.score(X_test, y_test))
```
